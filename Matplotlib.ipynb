### Matplotlib

#### Mathdatas plot library
- 메타데이터를 도표나 그래프(시각화)로 출력시키는 라이브러리
- 메타데이터 : 일정한 규칙에 따라 구조화, 표준화된 데이터
- 분석 결과를 수치만으로 설명하는 것보다 더 직관적으로 이해 가능
- 타인에게 분석 결과를 전달하기에 효과적

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### 플롯 차트(plot chart) : 선 그래프

```python
y = [2, 4, 6]

plt.plot(y) # 선 그래프 만들어주는 함수

plt.show() # 차트를 화면에 출력. (주피터 노트북에서는 생략해도 출력)
```

```python
# x, y 값 지정하여 선 그래프 그려보기
x = [1,2,3,4,5]
y = [1,5,7,8,6]

plt.plot(x,y, ls = '--', lw = 5, c = 'salmon', marker = 'D', mfc = 'skyblue', mec = 'indigo', ms = 10, mew = 2)
# ls : 선 스타일
# lw : 선 굵기
# c : 선 색깔
# marker : 마커의 종류
# mfc : 마커 색깔
# mec : 마커 테두리 선 색깔
# ms : 마커의 크기
# mew : 마커 테두리 선 굵기

plt.show()
```

```python
x = np.arange(7)
y = [1,9,9,3,9,9,1]

fig = plt.figure(figsize=(8,8), facecolor='red')
ax = fig.add_subplot()
ax.patch.set_facecolor('red')

plt.plot(x, y, ls = '-', lw = 50, c = 'yellow', marker = 'o', mfc = 'yellow', ms = 50)

ax.tick_params(axis='x', colors='red')
ax.tick_params(axis='y', colors='red')

plt.show()
```

### 그리드
- 차트의 선 위치들을 잘 보여주기 위해 차트 중간에 각자 선을 그어주는 것

```python
x = np.arange(7)
y = [1,4,6,8,9,5,3]

plt.plot(x, y, ls = '--', lw = 6, marker = 'o', ms = 15)
plt.grid()

plt.show()
```

### 틱(tick) 설정
- 플롯이나 차트에서, 축 상의 위치 표시 지점 : 틱(tick)
- 틱 위에 써진 숫자, 문자 : 틱 라벨(tick label)
- 자동으로 설정, 수동설정도 가능

```python
x = np.arange(7)
y = [1,4,6,8,9,5,3]

plt.plot(x, y, ls = '--', lw = 6, marker = 'o', ms = 15)
plt.grid()
plt.xticks([0,3,6])
plt.yticks([1,5,9])

plt.xlabel("X axis") # x축의 제목
plt.ylabel("Y axis") # y축의 제목
plt.title("X-Y graph") # 그래프 전체 제목

plt.show()
```

```python
x = np.arange(4)

y1 = [1,2,3,4]
y2 = [1,4,6,8]
y3 = [3,6,9,12]

plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.plot(x, y3, label='y3')

# 범례 설정 legend
plt.legend(loc='center left', prop = {'size' : 20})
# loc : 범례의 위치를 표시 (upper, center, lower) + (left, center, right)

plt.show()
```

### 바 차트(bar chart) : 막대그래프

```python
x = np.arange(6)
y = [80,85,70,60,50,90]

plt.bar(x,y, color = ['orange', 'mintcream', 'pink'], edgecolor='blue', lw = 3, alpha = 0.5, width = 0.5 )

# color : 바 내부 색상
# edgecolor : 바 테두리 색상
# linewidth(lw) : 테두리 너비
# alpha : 투명도 설정(0~1)
# width : 막대의 너비

plt.show()
```

```python
x = [0, 1, 2]
y = [100, 400, 850]
plt.yticks([200,400,600,800])
plt.bar(x,y, edgecolor='red', alpha = 0.5)
plt.show()
```
