```python
score_list = [['김미희', 89,78,99],
             ['서대희', 67,88,77],
             ['강예진', 100,55,96]]

score_list
```

```python
sum = 0
avg = 0
for i in range(len(score_list)):
    if(len(score_list[i]) < 5):
        for j in range(3):
            sum += score_list[i][j+1]
        avg = sum // 3
        score_list[i].append(avg)
        if (avg >= 90):
            score_list[i].append('A')
        elif (avg >= 80):
            score_list[i].append('B')
        elif (avg >= 70):
            score_list[i].append('C')
        elif (avg >= 60):
            score_list[i].append('D')
        else:
            score_list[i].append('F')
        sum = 0
avg_list = []
sum2 = 0
avg2 = 0
for i in range(3):
    for j in range(len(score_list)):
        sum2 += score_list[j][i+1]
    avg2 = sum2 // len(score_list)
    avg_list.append(avg2)
    sum2 = 0
print(avg_list)
print(score_list)
```

```python
print('======================================')
print('이름   Python  Java   DB   평균  학점')
print('======================================')
for i in range(len(score_list)):
    for j in range(6):
        if (score_list[i][j] == 100):
            print(score_list[i][j], end = '   ')
        else:
            print(score_list[i][j], end = '    ')
    print('\n')
print('======================================')
print(f'평균      {avg_list[0]}    {avg_list[1]}     {avg_list[2]}')
```
