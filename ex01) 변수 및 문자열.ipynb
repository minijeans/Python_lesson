#### 문자열 포맷팅
- 문자열 안의 특정한 값을 바꿔야할 경우 있을 때(변수)

```python
s1 = '오늘은 10월 21일 입니다.'

s1 = '오늘은 10월 22일 입니다.'

print(s1)
```

```python
# % 기호를 사용하는 방법

day = 21

# 정수형 변수를 넣을 자리에 %d, 마지막에 %(들어갈 변수명)

s1 = '오늘은 10월 %d일 입니다.'%day

# %d : 정수형
# %f : 실수형
# %s : 문자형

print(s1)
```

```python
day += 1
# day = day + 1

print('오늘은 10월 %d일 입니다.'%day)
```

```python
# format 함수를 사용한 포맷팅

month = 10
day = '오'

# .format(첫번째 변수명, 두번째 변수명..)

s = '오늘은 {}월 {}일 입니다.'.format(month, day)

print(s)
```

```python
# f문자열을 사용한 포맷팅
# 현재 가장 많이 쓰이는 방법, python3부터 제공하는 기능

month = 10
day = 21

s = f'오늘은 {month}월 {day}일 입니다.'

print(s)
```

#### 실습. 팀장의 이름을 입력받아 팀장의 성과 이름을 출력해보자.
- 입력된 값이 저장된 name에서 인덱싱, 슬라이싱
- 문자열 포맷팅

```python
name = input('팀장의 이름을 입력하세요: ')

print(f'팀장의 이름은 \'{name[1:]}\'이고, 성은 \'{name[0]}\'입니다.')
```
