### 리스트의 슬라이싱
- 인덱싱의 경우, 인덱싱 할 때 리스트 안의 요소
- 슬라이싱의 경우, 리스트의 형태로 가져옴

```python
list2 = [0,1,2,3,4]
```

```python
# [1, 2]
list2[1:3]
```

```python
# [0, 1]
list2[:2]
```

```python
# [3, 4]
list2[3:]
```

```python
# [3]
list2[3:4]
```

```python
list2[3]
```

```python
list3 = [1,2,3,['a','b','c']]

list3
```

```python
# list의 요소 변경 및 추가
list5 = [0,1,2]
print('변경 전', list5)

# [0,1,5]
list5[-1] = 5
print('변경 후', list5)

# [0,1,5,6]
# append()
# list의 요소를 추가
list5.append(6)
print('변경 후', list5)
```

```python
# 슬라이싱으로 요소 변경
list5 = [0,1,2,3,4,5,6]
print('변경 전 : ', list5)

list5[2:5] = ['a', 'b', 9, 9, 9, 9]
print('변경 후 : ', list5)
```

```python
array = [1, 2, 3, 4, 5]
```

```python
fruits = ['사과', '포도', ['수박','멜론'], '오렌지']
```

```python
apple = fruits[0]
melon = fruits[2][1]
orange = fruits[-1]

print(apple)
print(melon)
print(orange)
```

```python
choice_lst = []
choice_lst.append(fruits[0])
choice_lst.append(fruits[2][1])
choice_lst.append(fruits[-1])
print(choice_lst)
```

```python
# 한번에 값을 담아주는 함수
choice_lst2 = []

# extend : 리스트 안에 여러개의 값을 한번에 추가할 때
choice_lst2.extend([apple, melon, orange]) # list의 형태로 넣어준다.

choice_lst2
```

### 리스트의 다양한 함수들

```python
# 리스트의 길이(요소 개수)를 반환하는 함수
# len()
lst = [2,4,45,24,13,3,24,5,2,42,3,4,25,2,1]
print(len(lst))
```

```python
# 특정 값의 인덱스 반환
# index()
# list.index(찾고자하는 값)
# 가장 첫번째 나오는 값의 인덱스를 알려줌

lst.index(3)
```

```python
# 값을 추가할 때, 맨 끝이 아닌 내가 원하는 위치에 추가
# insert
# list.insert(인덱스위치, 값)

list5 = [0,1,2,3,4]
print('변경 전 :', list5)

list5.insert(2, [1,3,5])
print('변경 후 :', list5)
```

```python
# list 값 삭제
list7 = [33,44,55,66,44,77,44]
print('삭제 전 :', list7)

# 인덱스 번호로 삭제
# del
del list7[3]
print('삭제 후 :', list7)

# 값으로 삭제
# remove()
# list.remove(삭제할 값)

list7.remove(44)
print('44 삭제 후 :', list7) # while문과 in을 사용해서 44가 있는지 조건을 걸어서 풀면됨
```

```python
# 슬라이싱을 활용해서 삭제
list7 = [33,44,55,66,44,77,44,44,44,44,44,44]
print('삭제 전 :', list7)

# del
del list7[6:]
print('삭제 후 :', list7)
```

```python
# 오름차순 정렬
# sort()
# list.sort()

list8 = [6,3,5,8,2,1]
print('정렬 전 :', list8)

list8.sort()
print('오름차순 정렬 후 :', list8)
```

```python
# 역순
# reverse()
list8 = [6,3,5,8,2,1]
print('정렬 전 :', list8)

list8.reverse()
print('내림차순 정렬 후 :', list8)
```

```python
list8 = [6,3,5,8,2,1]

list8.sort()
list8.reverse()

list8
```
