### 실습! 하나의 셀에 코드 합치기!
1. 폴더 생성 자동화
2. 스크롤 하강
3. 링크 수집
4. 링크 기반으로 다운로드 -> tqdm 활용!
5. 자원반납

````python
# beautifulsoup과 selenium 합쳐서, soup은 quit이 중간에 있어도 됨 soup이라는 변수에 담겨 있기 때문에

import os
import time
from tqdm import tqdm # 
from urllib.request import urlretrieve # urlretrieve("이미지 링크", "경로 및 파일명.확장자")
from selenium import webdriver as wb # 테스트 브라우저 관리 도구
from selenium.webdriver.common.by import By # 경로 지정 방식
from selenium.webdriver.common.keys import Keys # 키보드 입력값
from bs4 import BeautifulSoup as bs # 파싱도구

# 1. 폴더 생성 자동화
if os.path.isdir("C:\\Users\\smhrd1\\Desktop\\포켓몬"):
    print("폴더가 이미 존재함")
else:
    os.mkdir("C:\\Users\\smhrd1\\Desktop\\포켓몬")
    print("폴더 생성 완료!")

# 2. 스크롤 하강
# 크롬 열고 최대화
driver = wb.Chrome()
url = "https://www.pokemonkorea.co.kr/pokedex#pokedex_1"
driver.get(url)

time.sleep(2) # 브라우저가 완전히 로드되게끔 대기

driver.maximize_window() # 최대화

# 페이지 스크롤 알고리즘!
while True:
    old = driver.page_source # 기존 페이지 소스

    # 스크롤 하강
    body = driver.find_element(By.TAG_NAME, "body")
    body.send_keys(Keys.END)
    time.sleep(2)
    
    new = driver.page_source # 갱신된 페이지 소스

    if old == new: # 같은지 확인
        print("스크롤 하강 완료")
        break
    else:
        pass

# 3. 링크 수집
soup = bs(driver.page_source, "lxml")
imgs = soup.select("img.img-fluid")

img_link = [img["src"] for img in imgs]

# 4. 링크 기반으로 다운로드
for i in tqdm(range(len(img_link))):
    urlretrieve(img_link[i], f"C:\\Users\\smhrd1\\Desktop\\포켓몬\\포켓몬{i+1}.jpg")

# 5. 자원 반납
driver.quit()
````

````python
# selenium, quit이 맨밑에 있어야 함
import os
import time
from tqdm import tqdm # 
from urllib.request import urlretrieve # urlretrieve("이미지 링크", "경로 및 파일명.확장자")
from selenium import webdriver as wb # 테스트 브라우저 관리 도구
from selenium.webdriver.common.by import By # 경로 지정 방식
from selenium.webdriver.common.keys import Keys # 키보드 입력값

# 1. 폴더 생성 자동화
if os.path.isdir("C:\\Users\\smhrd1\\Desktop\\포켓몬"):
    print("폴더가 이미 존재함")
else:
    os.mkdir("C:\\Users\\smhrd1\\Desktop\\포켓몬")
    print("폴더 생성 완료!")

# 2. 스크롤 하강
# 크롬 열고 최대화
driver = wb.Chrome()
url = "https://www.pokemonkorea.co.kr/pokedex#pokedex_1"
driver.get(url)
driver.maximize_window()

# 페이지 스크롤 알고리즘!
while True:
    old = driver.page_source # 기존 페이지 소스

    # 스크롤 하강
    body = driver.find_element(By.TAG_NAME, "body")
    body.send_keys(Keys.END)
    time.sleep(2)
    
    new = driver.page_source # 갱신된 페이지 소스

    if old == new: # 같은지 확인
        print("스크롤 하강 완료")
        break
    else:
        pass

# 3. 링크 수집
link_list = []
imgs = driver.find_elements(By.CSS_SELECTOR, "img.img-fluid")
for i in tqdm(range(len(imgs))):
        link_list.append(imgs[i].get_attribute("src"))

# 4. 링크 기반으로 다운로드
for i in tqdm(range(len(link_list))):
    urlretrieve(link_list[i], f"C:\\Users\\smhrd1\\Desktop\\포켓몬\\포켓몬{i+1}.jpg")

# 5. 자원 반납
driver.quit()
````
